13. Постройте график цены закрытия для Alphabet Inc. c 3 октября 2016 года по 7 октября 2016 года с отображением сетки (основная и второстепенная сетки) (с использованием minorticks_on).


```
fig, ax = plt.subplots(figsize=(5,4))
plt.xlabel('Date')
plt.ylabel('Closin value')
plt.title('Closing stock ...')
ax.set_ylim(772.5,777.0)
plt.plot(x, y, color='r', marker='o', linewidth=1)
plt.grid(color='r', linewidth=0.5, linestyle='-')
plt.minorticks_on()
plt.grid(which='minor', color='c', linestyle=':')
plt.show()
```
